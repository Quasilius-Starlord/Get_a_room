```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

# Change all non numerical value to integers in dataframe for Linear regression

```python
house_data['Property_Type'].unique()
```

```python
house_data['Property_Type'] = house_data['Property_Type'].replace(['Apartment', 'Bungalow', 'Single-family home', 'Duplex','Container Home'], [1,2,3,4,5])
```

```python
house_data['Furnishing'].unique()
```

```python
house_data['Furnishing'] = house_data['Furnishing'].replace(['Semi_Furnished', 'Unfurnished', 'Fully Furnished'], [1,2,3])
```

```python
house_data['Water_Supply'].unique()
```

```python
house_data['Water_Supply'] = house_data['Water_Supply'].replace(['Once in a day - Morning', 'Once in a day - Evening', 'All time','Once in two days'],[1,2,3,4])
```

```python
house_data['Crime_Rate'].unique()
```

```python
house_data['Crime_Rate'] = house_data['Crime_Rate'].replace(['Slightly below average', 'Well below average','Well above average', 'Slightly above average'], [1,2,3,4])
```

```python
house_data['Power_Backup'] = house_data['Power_Backup'].replace(['No', 'Yes', 'NOT MENTIONED'],[-1,1,0])
```

```python
house_data.head()
```

```python
house_data['Dust_and_Noise'].unique()
```

```python
house_data['Dust_and_Noise'] = house_data['Dust_and_Noise'].replace(['Medium', 'High', 'Low'], [2,5,1])
```

```python
house_data.head()
```

# Using Liner regression model

```python
train_data, test_data = train_test_split(house_data, test_size=0.15, random_state=10)
```

```python
train_data.columns
```

```python
X = train_data[['Property_Type', 'Property_Area', 'Number_of_Windows','Number_of_Doors', 'Furnishing', 'Frequency_of_Powercuts','Power_Backup', 'Water_Supply', 'Traffic_Density_Score', 'Crime_Rate','Dust_and_Noise', 'Air_Quality_Index', 'Neighborhood_Review']]
# X = train_data[['Property_Area']]
```

```python
Y = train_data[['Habitability_score']]
```

```python
model = linear_model.LinearRegression()
model.fit(X, Y)
```

# x is the testing dataframe and y is result dataframe

```python
x=test_data[['Property_Type', 'Property_Area', 'Number_of_Windows','Number_of_Doors', 'Furnishing', 'Frequency_of_Powercuts','Power_Backup', 'Water_Supply', 'Traffic_Density_Score', 'Crime_Rate','Dust_and_Noise', 'Air_Quality_Index', 'Neighborhood_Review']]
# x=test_data[['Property_Area']]
```

```python
y = test_data[['Habitability_score']]
```

```python
r=y.loc[y.index].values.tolist()
```

```python
p = model.predict(x.loc[x.index].values.tolist())
```

# Evaluation

```python
from sklearn.metrics import r2_score
```

```python
r2_score(r, p)
```
